```python
"""
使用梯度下降方法求函数loss=x^2的极小值
梯度下降方法：
1.假设初始点是x0,计算f'(x0)
2.更新x1=x0-f'(x0)*deta
3.重复上述步骤
"""
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(-3, 3, 0.01)
y = x ** 2
```

```python
plt.plot(x, y)
```

```python
step = 1000
rate = 0.1
result = []
_x = -2
for i in range(step):
    _x = _x - (2 * _x * rate)
    result.append([_x, _x ** 2])
result = np.array(result)
plt.plot(x, y)
plt.scatter(result[:, 0], result[:, 1], c="red")
```

```python
"""
批数据中，方程是由批数据描述:y1=wx1+b+loss,y2=wx2+b+loss...
希望是结果是loss=Sum((WX+B-Y)^2)，取得最小值
其中：y相当于标签，x和b相当于学习出来的参数，loss就相当于实际-预测
"""
def fx(w, b):
    def inner_fx(x):
        return w * x + b
    return inner_fx

f = fx(2, 1)
x = np.random.random(size=100)
y = f(x) + np.random.random(size=100) * 0.5
plt.scatter(x, y)
```

```python
# 数据集梯度下降
w = 0
b = 0
step = 1000
rate = 0.5
for i in range(step):
    _w = 0
    _b = 0
    for j in range(len(x)):
        _w += 2 * x[j] * (w * x[j] + b - y[j])
        _b += 2 * (w * x[j] + b - y[j])
    # 注意求出的_w和_b是梯度的len(x)倍，所以更新前需要除以len(x)
    w = w - rate * _w / len(x)
    b = b - rate * _b / len(x)
print(w, b)
plt.scatter(x,y)
f=fx(w,b)
plt.plot(np.arange(0,1,0.05),f(np.arange(0,1,0.05)),c="red")
```

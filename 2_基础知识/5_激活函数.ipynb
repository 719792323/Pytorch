```python
import torch
from matplotlib import pyplot as plt
```

```python
# relu函数
# relu(x)=max(x,0)，即小于等于0的值都为0，大于0的值为本身
x=torch.arange(-5,5,0.1,requires_grad=True)
y=torch.relu(x)
#如果x设置要求梯度，那么不能直接用于绘图，因为有梯度信息，系统认为这样不安全x和y都要调用detach()
# 否则会报错
plt.plot(x.detach(),y.detach())
```

```python
# 绘制relu的导数图像
# 因为上述计算导致y的计算结果是标量，要求分量梯度需要加torch.ones_like(x)
# 注意如果还需要反复运行关于x的动态图，需要加retain_graph=True，否则运行完backward后
# pytorch会摧毁动态图信息以节省内存
# retain_graph=True只对单次backward有效，如果下一个backward没加依然会摧毁动态图
x.grad.zero_()
y.backward(torch.ones_like(x),retain_graph=True)
plt.plot(x.detach(),x.grad)
```

```python
# sigmoid函数
# sigmoid(x)=1/(1+exp(-x))
y=torch.sigmoid(x)
plt.plot(x.detach(),y.detach())
```

```python
# sigmoid导数图像
x.grad.zero_()
y.backward(torch.ones_like(x),retain_graph=True)
plt.plot(x.detach(),x.grad)
```

```python
# tanh(x)函数
# tanh(x)=(1-exp(-2x))/(1+exp(-2x))
y=torch.tanh(x)
plt.plot(x.detach(),y.detach())
```

```python
# tanh函数导数图像
x.grad.zero_()
y.backward(torch.ones_like(x),retain_graph=True)
plt.plot(x.detach(),x.grad)
```
